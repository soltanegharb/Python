```python
#load libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier #Import Decision Tree Classifier
from sklearn.model_selection import train_test_split #Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
```

```python
#Working with the dataset here

col_names = ['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal','target']
data = pd.read_csv("heart.csv",header=None, names=col_names)
```

```python
data
```

```python
#split the dataset in features and target variable
feature_cols = ['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal']
x = data[feature_cols]
x = data.iloc[1:,:-1]
y = data['target']
y = data.iloc[1:,-1]
```

```python
#Splitting data into Train and Test Sets
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
#Create Decision Tree Classifier object using entropy
clf_entropy = DecisionTreeClassifier(criterion="entropy", max_depth=2)

#Train Decision Tree Classifier
clf_entropy = clf_entropy.fit(x_train, y_train)
```

```python
#Predict the response for test dataset
y_pred = clf_entropy.predict(x_test)
#user_input = [55,1,0,160,289,0,0,145,1,0.8,1,1,3]
#y_pred2 = clf_entropy.predict(user_input)
print("Accuracy", metrics.accuracy_score(y_test,y_pred))
```

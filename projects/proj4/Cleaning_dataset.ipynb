```python
#Import the libraries
import numpy as np
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder,StandardScaler

#Read the dataset
data = pd.read_csv('toy_dataset.csv') #data stays untouched just in case we need it

edited_data = pd.read_csv('toy_dataset.csv')
```

```python
#Finding Null cells
edited_data.isnull().sum()
```

```python
#Anomaly Detection  with z_score
mean_age = edited_data.iloc[:,3].mean()
std_age = edited_data.iloc[:,3].std()

outliers_age=[]
outliers_age_test=[]

for num , item in zip(edited_data.iloc[:,0],edited_data.iloc[:,3]):
    z_score = (item - mean_age) / std_age
    if z_score > 3:
        outliers_age.append([num,item])
        edited_data.loc[num-1,'Age']=int(mean_age.round())

mean_income = edited_data.iloc[:,4].mean()
std_income = edited_data.iloc[:,4].std()

outliers_income=[]
outliers_income_test=[]

for num , item in zip(edited_data.iloc[:,0],edited_data.iloc[:,4]):
    z_score = (item - mean_income) / std_income
    if z_score > 3:
        outliers_income.append([num,mean_income.round(),item])
        edited_data.loc[num-1,'Income']=mean_income.round()

print("number of outliers in age column before pre processing", len(outliers_age),"\n")
print(outliers_age)
print("\n\n\n")
print("number of outliers in income column before pre processing", len(outliers_income),"\n")
print(outliers_income)
```

```python
#Test the anomaly detection on edited_data

outliers_age_test=[]
for num , item in zip(edited_data.iloc[:,0],edited_data.iloc[:,3]):
    z_score = (item - mean_age) / std_age
    if z_score > 3:
        outliers_age_test.append([num,mean_age,item])


mean_income = edited_data.iloc[:,4].mean()
std_income = edited_data.iloc[:,4].std()

outliers_income_test=[]
for num , item in zip(edited_data.iloc[:,0],edited_data.iloc[:,4]):
    z_score = (item - mean_income) / std_income
    if z_score > 3:
        outliers_income_test.append([num,mean_income.round(),item])

print("number of outliers in Age column after test", len(outliers_age_test),"\n")
print("number of outliers in Income column after test", len(outliers_income_test),"\n")
```

```python
#Scale the edited_data

edited_data.iloc[:,3] = (edited_data.iloc[:,3] - edited_data.iloc[:,3].min()) / (edited_data.iloc[:,3].max() - edited_data.iloc[:,3].min())
print(edited_data.iloc[:,3])

edited_data.iloc[:,4] = (edited_data.iloc[:,4] - edited_data.iloc[:,4].min()) / (edited_data.iloc[:,4].max() - edited_data.iloc[:,4].min())
print(edited_data.iloc[:,4])
edited_data
```
